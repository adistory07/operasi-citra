```python
from PIL import Image

def clipping(intensitas):
  if intensitas < 0:
    return 0
  if intensitas > 255:
    return 255
  return intensitas

def atur_pencerahan(nilai_pencerahan, nama_setelah_disave):
  CITRA = Image.open('pictogram-884043_1280.PNG')
  PIXEL = CITRA.load()

  ukuran_horizontal = CITRA.size[0]
  ukuran_vertikal = CITRA.size[1]

  # The for loops should be indented to be part of the function
  for x in range(ukuran_horizontal):
    for y in range(ukuran_vertikal):
      R = clipping(PIXEL[x, y][0] + nilai_pencerahan)
      G = clipping(PIXEL[x, y][1] + nilai_pencerahan)
      B = clipping(PIXEL[x, y][2] + nilai_pencerahan)
      PIXEL[x, y] = (R, G, B)

  CITRA.save(nama_setelah_disave)

atur_pencerahan(128, 'pictogram-884043_1280_dicerahkan.PNG')
atur_pencerahan(-128, 'pictogram-884043_1280_digelapkan.PNG')
```

```python
# Import library yang dibutuhkan
from PIL import Image
from PIL import ImageEnhance
# Buka file gambar
image = Image.open('disabilitas.png')
# Konversi gambar ke mode RGB
image = image.convert('RGB')  # Convert to RGB mode
# Menampilkan gambar di penampil gambar
image.show()
# Meningkatkan kecerahan
curr_bri = ImageEnhance.Brightness(image)
# Kecerahan ditingkatkan dengan faktor pengali 1.5
new_bri = 1.5
img_brightened = curr_bri.enhance(new_bri)
# Menampilkan gambar update yang sudah di-adjust tingkat kecerahannya
img_brightened.save("terang_disabilitas.png")
```

```python
# Import library yang diperlukan
from PIL import Image
from PIL import ImageEnhance
# Buka file gambar
image = Image.open('disabilitas.png')
# Konversi gambar ke mode RGB
image = image.convert('RGB')  # Convert to RGB mode
# Menampilkan gambar original
image.show()
# Meningkatkan tingkat kewarnaan
curr_col = ImageEnhance.Color(image)
# Meningkat tingkat kewarnaan dengan faktor 2.0
new_col = 2.0
img_colored = curr_col.enhance(new_col)
# Menampilkan gambar update yang sudah di-adjust tingkat kewarnaan
img_colored.save("disabilitas_bewarna.png")
```

```python
# Import library yang diperlukan
from PIL import Image
from PIL import ImageEnhance
# Buka file gambar
image = Image.open('disabilitas_bewarna.png')
# Menampilkan gambar
image.show()
# Meningkatkan kontras
curr_con = ImageEnhance.Contrast(image)
# Peningkatan kontras dengan faktor 0.3
new_con = 0.3
img_contrasted = curr_con.enhance(new_con)
# Menampilkan gambar update yang sudah di-adjust kontras nya
img_contrasted.save("disabilitas_kontras.jpg")
```
