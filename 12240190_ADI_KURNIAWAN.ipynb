```python
from PIL import Image

def clipping(intensitas):
  if intensitas < 0:
    return 0
  if intensitas > 255:
    return 255
  return intensitas

def atur_pencerahan(nilai_pencerahan, nama_setelah_disave):
  CITRA = Image.open('pictogram-884043_1280.PNG')
  PIXEL = CITRA.load()

  ukuran_horizontal = CITRA.size[0]
  ukuran_vertikal = CITRA.size[1]

  # The for loops should be indented to be part of the function
  for x in range(ukuran_horizontal):
    for y in range(ukuran_vertikal):
      R = clipping(PIXEL[x, y][0] + nilai_pencerahan)
      G = clipping(PIXEL[x, y][1] + nilai_pencerahan)
      B = clipping(PIXEL[x, y][2] + nilai_pencerahan)
      PIXEL[x, y] = (R, G, B)

  CITRA.save(nama_setelah_disave)

atur_pencerahan(128, 'pictogram-884043_1280_dicerahkan.PNG')
atur_pencerahan(-128, 'pictogram-884043_1280_digelapkan.PNG')
```

```python
# Filter linier
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
image = cv.imread("disabilitas.png")
fig, ax = plt.subplots(1, 3, figsize=(16, 8))
fig.tight_layout()
# untuk meng-konvolusi kernel pada sebuah gambar kita dapat menggunakan cv.filter2D
kernel_sharpening = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
kernel_sharpening_2 = np.array([[-1, -1, -1], [-1, 10, -1], [-1, -1, -1]])

ax[0].imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
ax[0].set_title('Original Image')
sharpened = cv.filter2D(image, -1, kernel_sharpening)
ax[1].imshow(cv.cvtColor(sharpened, cv.COLOR_BGR2RGB))
ax[1].set_title('Sharpened Kernel Image')
sharpened_2 = cv.filter2D(image, -1, kernel_sharpening_2)
ax[2].imshow(cv.cvtColor(sharpened_2, cv.COLOR_BGR2RGB))
ax[2].set_title('Sharpened Kernel Image 2')
plt.show()
```

```python
# Import library yang diperlukan
from PIL import Image
from PIL import ImageEnhance

# Membuka file gambar
image = Image.open('disabilitas.png')

# Convert the image to RGB mode if it's in palette mode
if image.mode == 'P':
    image = image.convert('RGB')

# Menampilkan gambar
image.show()

# Meningkatkan ketajaman
curr_sharp = ImageEnhance.Sharpness(image)

# Peningkatan ketajaman dengan faktor 50
new_sharp = 50
img_tajam = curr_sharp.enhance(new_sharp)

# Menampilkan gambar update yang sudah di-adjust ketajamannya
img_tajam.save("disabilitas_kereng.png")
```
